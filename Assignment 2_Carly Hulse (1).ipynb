```python
# Reading in the data set:
import pandas as pd
## import numpy as np

# Note that missing data is represented by ?s! Would rather have them as NaN.
data = pd.read_csv('adultsData.csv', index_col=0, na_values=["?"], skipinitialspace=True)# have to skip initial space because '?'s are actually ' ?'s in excel file

# Now replace NaN's with modes for all columns
modes = data.mode()
values = modes.iloc[0,]
data.fillna(value=values, inplace=True)

data.head(15) # check that reading in the data correctly including ? in row 14 - yep!
```

```python
# 1. How many men and women are represented in this dataset?
data["sex"].value_counts() # There are 21790 males, 10771 females, and no missing values in the sex column of this dataset.
```

```python
# 2. What is the average age of women?
data.groupby('sex').age.mean()['Female'] # the average age of women is 36.86

print("The average age of women is", round(data.groupby('sex').age.mean()['Female'],2))
```

```python
# 3. What is the percentage of German citizens and Canadian citizens?
percents = data['native-country'].value_counts(normalize=True) * 100
    # this pulls all of the countries, counts them
    # normalize=True gives relative frequencies of the unique values
    # *100 gives percentage

print("The percentage of German citizens is", round(percents['Germany'],2), "% and the percentage of Canadian citizens is", round(percents['Canada'],2),"%")
```

```python
# 4. What are the mean and standard deviation of the age for those who earn more than 50K per year and
# those who earn less than 50K per year?

morethan = data.loc[data['salary'] == '>50K', 'age']
lessthan = data.loc[data['salary'] == '<=50K', 'age']

print("The mean of age for those who earn more than 50k per year is", round(morethan.mean()), "years with a standard deviation of", round(morethan.std(),1), "years.")

print("The mean of age for those who earn less than 50k per year is", round(lessthan.mean()), "years with a standard deviation of", round(lessthan.std(),1), "years.")
```

```python
# 5. What is the education level of people who earn more than 50K? Is it true that they have at least high school education?
data.loc[data['salary'] == '>50K', 'education'].unique()

print("People who earn more than 50K have the following eduaction levels:", (data.loc[data['salary'] == '>50K', 'education'].unique()), "\nWe can see that not all have high school education; for instance, some just have 7th or 8th grade.")
```

```python
# 6. Display age statistics for each race and each gender. Use groupby() and describe(). 
# Find the maximum age of men and women in each race group.

#print(data.groupby('race').age.describe()) # age statistics for each race
#print(data.groupby('sex').age.describe()) # age statistics for each gender

print(data.groupby(['race', 'sex']).age.describe()) # age statistics for each race and each gender

# max age
print("\n\nMax age of men and women in each race group:\n", data.groupby(['race', 'sex']).age.max())
```

```python
# 7. Among whom is the proportion of those who earn greater than 50K: married or single?
# Consider as married those who have a marital-status starting with Married 
# (e.g., Married-civ-spouse, Married-spouse-absent, Married-AF-spouse, etc.).

greaterthan = data.loc[data['salary'] == '>50K', 'marital-status']

married = greaterthan.str.contains('Married-')

married.value_counts()

single = married.value_counts().loc[False]
mar = married.value_counts().loc[True]

singlenorm = married.value_counts(normalize=True).loc[False]
marnorm = married.value_counts(normalize=True).loc[True]

print("The ratio of those who earn greater than 50k that are married to those that are single is", mar, ":", single, "\nThis corresponds to", round(marnorm,2), "and", round(singlenorm,2), "of the total", greaterthan.size, "people earning over 50K, respectively." )
```

```python
# 8. What is the maximum number of hours a person works per week? 
# How many people work such a number of hours, and what is the percentage of those who earn more than 50K among them?

print("The maximum number of hours a person works per week is", data['hours-per-week'].max(), "hours.")

print(sum(data['hours-per-week'] == (data['hours-per-week'].max())), "people work the max number of weekly hours.")

earn50k = ((data['salary'] == '>50K') & (data['hours-per-week'] == (data['hours-per-week'].max()))).value_counts().loc[True]
percent = round(earn50k/(sum(data['hours-per-week'] == (data['hours-per-week'].max()))) * 100)

print("For those who work the max number of hours,", percent,"% make more than 50K.")
```
